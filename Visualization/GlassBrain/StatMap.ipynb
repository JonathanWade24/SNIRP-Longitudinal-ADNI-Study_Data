```python
import numpy as np
import nibabel as nib

atlas_img = nib.load("C:/Users/jonat/Downloads/cat12_neuromorphometrics.nii")
atlas_data = atlas_img.get_fdata()
stat_map = np.zeros_like(atlas_data)
```

```python
filtered_df = pd.read_csv('C:/Users/jonat/Downloads/WM_fixed.csv')

for _, row in filtered_df.iterrows():
    roi_name = row['Outcome_Variable']
    estimate = row['estimate']  # Assuming this is the name of your column with the statistical measure
    if roi_name in roi_mapping:
        index = roi_mapping[roi_name]
        # Fill in the region in the statistical map
        stat_map[atlas_data == index] = estimate

print(filtered_df[['Outcome_Variable', 'estimate']])

print(roi_mapping.get('Right Amygdala'))
```

```python
from nilearn import plotting
import matplotlib.pyplot as plt

# Assuming 'adjusted_stat_map_img' is your NIfTI image with the correct statistical data
# Check the data range and standard deviation to understand the spread of values
data = stat_map_img.get_fdata()
data_range = [data.min(), data.max()]
std = data.std()

print('Data range:', data_range)
print('Standard deviation:', std)

# If the data range is too narrow, consider adjusting the color bar limits to be within one standard deviation of the mean
vmin, vmax = data_range[0], data_range[1]  # or for example, mean +/- 1 * std

# Now plot the glass brain with the specified `vmin` and `vmax`
display = plotting.plot_glass_brain(
    stat_map_img,
    display_mode='ortho',
    colorbar=True,
    cmap='Reds',  # This is a diverging colormap suitable for both positive and negative values
    vmin=vmin,  # Manually setting the minimum value for color mapping
    vmax=vmax,  # Manually setting the maximum value for color mapping
    symmetric_cbar='auto',  # Automatically determines if the color bar should be symmetrical or not
    title="Glass Brain Visualization Including Negative Estimates",
    plot_abs= False
)

# Show the plot
plotting.show()
```
